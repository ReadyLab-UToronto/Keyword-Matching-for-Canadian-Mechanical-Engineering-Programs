```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
from selenium import webdriver
import time
import re
```

```python
url = "https://www.westerncalendar.uwo.ca/Modules.cfm?ModuleID=21609&SelectedCalendar=Live&ArchiveID="
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--incognito')
chrome_options.add_argument('--headless')

driver = webdriver.Chrome("C:\\Users\\jerry\\Downloads\\chromedriver", options=chrome_options)
```

```python
driver.get(url)
```

# 1. Collect all course link texts for webdriver to click on

```python
page_soup = soup(driver.page_source, 'lxml')
```

```python
blocks = page_soup.find("div", {"class": "moduleInfo"}).findAll("p")
len(blocks)
```

```python
core_blocks = blocks[:6]
elective_block_start = blocks[6].find("strong", text=" Biomedical Engineering (BME) Technical Electives (Courses that do not appear on this list require special permission):")
elective_block_start
```

```python
core_containers = [container for block in core_blocks for container in block.findAll("a")]
core_link_texts = [container.text.strip() for container in core_containers]
core_link_texts
```

```python
len(core_link_texts)
```

```python
elec_containers = []
iterator = elective_block_start.findNext()
while iterator.text != " Non-technical Electives:":
    elec_containers.append(iterator)
    print(iterator.text)
    iterator = iterator.findNext()
```

```python
elec_containers
```

```python
elec_link_texts = [container.text.strip() for container in elec_containers if len(container.text)>0]
elec_link_texts
```

```python
link_texts = core_link_texts + elec_link_texts
link_texts
```

# 2. Automation script to scrape all courses

```python
from selenium.common.exceptions import NoSuchElementException

course_codes = []
course_names = []
course_descs = []
counter = 0
    
for link_text in link_texts:
    #go to course page
    try:
        link = driver.find_element_by_link_text(link_text)
    except NoSuchElementException:
        print("no link for {}".format(link_text))
        continue

    time.sleep(2)
    link.click()    
    time.sleep(2)
    
    #scrape course info
    page_soup = soup(driver.page_source, 'lxml')
    course_info = page_soup.find("div", {"id": "CourseInformationDiv"})
    course_code = course_info.find("h2").text
    course_name = course_info.find("h3").text
    course_desc = course_info.findAll("div", {"class": None, "id": None})[1].text.strip()
    
    course_codes.append(course_code)
    course_names.append(course_name)
    course_descs.append(course_desc)
    
    print("Scraped ", course_codes[-1])
    counter += 1
    
    #go to course list page
    driver.back()
    time.sleep(2)
        
print("Finished scraping {} courses".format(counter))
```
