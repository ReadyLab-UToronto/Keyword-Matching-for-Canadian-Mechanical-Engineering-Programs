```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
from selenium import webdriver
import time
import re
```

# All possible Engineering courses to choose as the free elective

```python
url = "https://www.mcgill.ca/study/2020-2021/faculties/engineering/undergraduate/courses/engineering/undergraduate/engineering/undergraduate?sort_by=field_subject_code&f%5B0%5D=level%3Aundergraduate&f%5B1%5D=field_faculty_code%3AEN&f%5B2%5D=course_level%3A300&f%5B3%5D=course_level%3A400&f%5B4%5D=course_level%3A500"
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--incognito')
#chrome_options.add_argument('--headless')

driver = webdriver.Chrome("C:\\Users\\jerry\\Downloads\\chromedriver", options=chrome_options)
```

```python
driver.get(url)
```

```python
page_soup = soup(driver.page_source, 'lxml')
```

# 1. Collect course link texts on the current page

```python
containers = page_soup.find("div", {"class": "view-content"}).findAll("a")
len(containers)
```

```python
link_texts = [container.text.strip().replace("\n", "").replace("  ", " ") for container in containers]
link_texts
```

# 2. Automation script for scraping Engineering level 300 - 500 courses

```python
from selenium.common.exceptions import NoSuchElementException

course_names = []
course_codes = []
course_descs = []
counter = 0

for link_text in link_texts:
    try:
        link = driver.find_element_by_link_text(link_text)
    except NoSuchElementException:
        print("no link for {}".format(link_text))
        continue
        
    time.sleep(1)
    link.click()    
    time.sleep(3)
    page_soup = soup(driver.page_source, 'lxml')
    
    try:
        course_title = re.split(" \([0-9]+ credit[s]*\)", page_soup.find("h1", {"id": "page-title"}).text.strip())[0]
    except AttributeError:
        waiting = input("Internet connection possibly lost. Ready to continue?")
        course_title = re.split(" \([0-9]+ credit[s]*\)", page_soup.find("h1", {"id": "page-title"}).text.strip())[0]
        
    course_codes.append(" ".join(course_title.split()[:2]))
    course_names.append(" ".join(course_title.split()[2:]))
    course_descs.append(page_soup.find("h3", text = "Overview").findNextSibling().text.strip())
    
    print("Scraped ", course_codes[-1])
    counter += 1    
    
    driver.back()
    time.sleep(3)
    
    if link_text == link_texts[-1]:
        try:
            driver.find_element_by_link_text("‚ùØ").click()
        except NoSuchElementException:
            print("end has been reached")
            break
            
        time.sleep(3)
        page_soup = soup(driver.page_source, 'lxml')
        
        containers = page_soup.find("div", {"class": "view-content"}).findAll("a")
        link_texts += [container.text.strip().replace("\n", "").replace("  ", " ") for container in containers]

        
print("Finished scraping {} courses".format(counter))
```
