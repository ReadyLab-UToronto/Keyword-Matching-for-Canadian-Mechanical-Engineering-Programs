```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
from selenium import webdriver
import time
import re
```

```python
url = "https://calendar.carleton.ca/undergrad/courses/"
```

```python
#electives for mech eng: "4000-level Mechanical and Aerospace Engineering (MAAE, AERO or MECH)"
course_types = ["AERO", "MAAE", "MECH"]
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--incognito')
#chrome_options.add_argument('--headless')

driver = webdriver.Chrome("C:\\Users\\jerry\\Downloads\\chromedriver", options=chrome_options)
```

```python
driver.get(url)
```

# 1. Automation script to scrape all 4000-level MAAE, AERO, MECH courses

```python
page_soup = soup(driver.page_source, 'lxml')
```

```python
course_codes = []
course_names = []
course_descs = []

for course_type in course_types:
    
    counter = 0
    
    driver.get(url)
    time.sleep(3)
    driver.find_element_by_link_text(course_type).click()
    time.sleep(3)
    
    page_soup = soup(driver.page_source, 'lxml')
    containers = page_soup.findAll("div", {"class": "courseblock"})
    
    for container in containers:
        #look for course codes that start with 4 (4000s level)
        if container.find("span", {"class": "courseblockcode"}).text.replace("\xa0", " ").strip().split()[1][0] == '4':
            course_code = container.find("span", {"class": "courseblockcode"}).text.replace("\xa0", " ").strip()
            course_name = container.find("span", {"class": "courseblocktitle"}).text.split("\n")[1].strip()
            course_desc = container.strong.next_sibling.next_sibling.strip()
            
            course_codes.append(course_code)
            course_names.append(course_name)
            course_descs.append(course_desc)
            
            print("Scraped ", course_codes[-1])
            counter += 1
    
    print("Finished scraping {} {} courses".format(counter, course_type))
    print("")
```
